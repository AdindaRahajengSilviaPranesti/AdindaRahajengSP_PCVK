data types?
- Answer = Int16 is suitable for situations where storage space is an issue and lower levels of precision are acceptable, while int32 is better used when high precision and detail in black images is required although it results in larger file sizes and int32 is a better choice for avoiding overflow and data loss.
```python
import cv2
import numpy as np
import random
from google.colab.patches import cv2_imshow

# Load the image
image_path = '/content/drive/MyDrive/kitten_Adinda/kitten1.jpeg'
img = cv2.imread(image_path)

# Create an output image to draw the boxes on
output_img = img.copy()

# Create 5 boxes on the image
for _ in range(5):
    color = (random.randint(0, 255), random.randint(0, 255), random.randint(0, 255))  # Random color
    x1, y1 = random.randint(0, img.shape[1]), random.randint(0, img.shape[0])  # Random top-left coordinates
    x2, y2 = x1 + random.randint(10, 100), y1 + random.randint(10, 100)  # Random bottom-right coordinates

    thickness = random.randint(1, 5)  # Random thickness
    cv2.rectangle(output_img, (x1, y1), (x2, y2), color, thickness)

# Display the image with boxes
cv2_imshow(output_img)
cv2.waitKey(0)
cv2.destroyAllWindows()

```
