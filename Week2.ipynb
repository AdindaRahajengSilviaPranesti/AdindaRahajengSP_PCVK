```python
cv.rectangle(black_img,pt1=(200,200),pt2=(300,300),color=(0,0,255),thickness=15)
plt.imshow(black_img)
```

```python
black_img2=np.zeros(shape=(512,512,3),dtype=np.int32)
plt.imshow(black_img2)
```

```python
vertices = np.array([[100,300],[200,200],[400,300],[200,400]],dtype=np.int32)
vertices
```

```python
pts = vertices.reshape((-1,1,2)) #nilai 2 untuk menunjukkan bahwa tiap titik dibuat 3 channel yg mewakili
pts
```

```python
cv.polylines(black_img2,[pts],isClosed=True,color=(255,0,0),thickness=5)
plt.imshow(black_img2)
```

2. What is the difference and effect of creating a black image between int16 and int32
data types?
- Answer = Int16 is suitable for situations where storage space is an issue and lower levels of precision are acceptable, while int32 is better used when high precision and detail in black images is required although it results in larger file sizes and int32 is a better choice for avoiding overflow and data loss. The choice of this data type depends on the specific needs of the project and how critical precision is in the representation of gray levels in black images.
